**1. Data Exploration**

```python
import pandas as pd
import numpy as np #numerical_operation
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv ('Lesson_4_student_grades_dataset.csv')
print(data.head())
```

The features above has the potential impact on the grades of a student, wherein it depends to the data if the student would have a high or low grades.

**2. Data Preprocessing**



```python
print(data.isnull().sum())
```

Normalizing or standardizing numerical features is important as it helps you identify if the data are complete or if something is/are missing.

3. **Train-Test Split**

```python
x = data.drop('Grades', axis=1)
y = data ['Grades']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(x_train,y_train)
```

**4. Linear Regression Model**

```python
x = data [['Hours_of_Study']]
y = data [['Grades']]

model = LinearRegression()
model.fit(x,y)

plt.scatter(x,y)
plt.plot(x,model.predict(x), color='pink')
plt.xlabel('Hours_of_Study')
plt.ylabel('Grades')
plt.title('Linear Regression')
plt.show
```
